# Plot Growth Curve - single measure, without using biological replicates 

```r
## Load Packages
library(ggplot2) 
library(patchwork)
```

```r
## Import OD tables (one tsv table per isolate)
df_test <- read.delim("/home/joao/Documents/Research/Mestrado/github/data/df_test.tsv") 
df_testt <- read.delim("/home/joao/Documents/Research/Mestrado/github/data/df_testt.tsv") 
head(df_test)
head(df_testt)
```

```r
## Plot growth curves - one per isolate

### Isolate 1
plot1 <- ggplot(df_test, aes(x = Time, y = OD, color = Condition)) + # specify your dataframe, x and y axis 
    geom_line(size = 1, linetype = "solid") +  # Use solid lines for all conditions
    geom_point(size = 2) +                    # Points retain the color of the condition
    labs(
        title = "Growth Curve for a Bacterial Isolate 1", # change depending on what's the plot about
        x = "Time (h)",
        y = "Optical Density (OD600)"
    ) +
    theme_minimal() +
    theme(
        plot.title = element_text(hjust = 0.5),
    ) +
    scale_color_manual(values = c("GLU" = "lightblue", "SIA" = "pink", "NC" = "gray")) # Set specific colors for conditions

plot1 # Visualize the individual plot

### Isolate 2
plot2 <- ggplot(df_testt, aes(x = Time, y = OD, color = Condition)) + # specify your dataframe, x and y axis 
    geom_line(size = 1, linetype = "solid") +  # Use solid lines for all conditions
    geom_point(size = 2) +                    # Points retain the color of the condition
    labs(
        title = "Growth Curve for a Bacterial Isolate 2", # change depending on what's the plot about
        x = "Time (h)",
        y = "Optical Density (OD600)"
    ) +
    theme_minimal() +
    theme(
        plot.title = element_text(hjust = 0.5),
    ) +
    scale_color_manual(values = c("GLU" = "lightblue", "SIA" = "pink", "NC" = "gray")) # Set specific colors for conditions

plot2 # Visualize the individual plot
```

```r
## Combine all plots in a single panel
combined_plot <- plot1 + plot2 + plot_layout(ncol = 2, nrow = 1) # specify number of rows and columns  as needed

combined_plot # Visualize it
```
